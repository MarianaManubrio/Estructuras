## 1. Resuelve los ejercicios utilizando el mapa definido a continuación.

a) Crea un mapa que contenga los datos del archivo paises.csv en el año 2013
    - Clave: nombre del país
    - Valor: La expecativa de vida

```python
import csv
from mapas import mapa
file = open('paises.csv')
lector = csv.reader(file)
lista = [x for x in lector if x[2] == "2013"]
mapa2013 = mapa()
for elemento in lista:
    nombre = str(elemento[0])
    expectativa = float(elemento[1])
    mapa2013[nombre] = expectativa
```

b) Obtén y muestra la expectativa de vida en México, Guatemala, Belize, El Salvador y Nicaragua.

```python
nombres = []
for i in range(len(mapa2013.tabla)):
    if mapa2013.tabla[i].llave == "Mexico":
        nombres.append(mapa2013.tabla[i])
    if mapa2013.tabla[i].llave == "Guatemala":
        nombres.append(mapa2013.tabla[i])
    if mapa2013.tabla[i].llave == "Belize":
        nombres.append(mapa2013.tabla[i])
    if mapa2013.tabla[i].llave == "El Salvador":
        nombres.append(mapa2013.tabla[i])
    if mapa2013.tabla[i].llave == "Nicaragua":
        nombres.append(mapa2013.tabla[i])
print("b) Expectativa de vida en Mexico, Guatemala, Belize, El Salvador y Nicaragua:")
for item in nombres:
    print(item.llave, item.valor)
```

c) Encuentra la expectativa de vida para cada uno de los paises anteriores.

```python
for i in range(len(mapa2013.tabla)):
    print(mapa2013.tabla[i].llave, ": ", mapa2013.tabla[i].valor)
```

d) Obten la lista de todas las claves del mapa.

```python
claves = []
for node in mapa2013.tabla:
    claves.append(node.llave)
print("d) Lista de todas las claves del mapa: ")
for clave in claves:
    print(clave)
```

e) Filtra la lista anterior para mostrar sólo los paises que empiezan con "C":

```python
result = [x for x in claves if x[0] == 'C']
print("e) Lista anterior filtrada para mostrar sólo los países que empiezan con C:")
for item in result:
    print(item)
```

f) Calcula el promedio de expectativa de vida para todos los paises que empiezan
con C.

```python
sum1 = 0
listaNodos = []
index = 0
for item in mapa2013.tabla:
    if item.llave[0] == 'C':
        listaNodos.append(mapa2013.tabla[index])
    index += 1

for item in listaNodos:
    sum1 += float(item.valor)
average = sum1/len(listaNodos)

print("f) Promedio de expectativa de vida para todos los paises que empiecen con C:", average)
```
